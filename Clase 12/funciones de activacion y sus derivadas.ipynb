# Deep Learning Activation functions

Las funciones de activación son una parte importante de las redes neuronales. Introducen no linealidades que ayudan a añadir más expresividad a las redes neuronales. En las redes de vainilla se ha utilizado la función Sigmoid. Sin embargo, tiene algunos problemas con la velocidad de convergencia y la saturación del gradiente. Por lo tanto, se han propuesto diferentes fucniones de activación para acelerar la convergencia, aliviar la saturación del gradiente y mitigar el desplazamiento de las covariables. En este cuaderno exploraremos algunas de las funciones de activación alternativas más populares utilizadas en las arquitecturas de aprendizaje profundo.


```python
import numpy as np
import matplotlib.pyplot as plt

def approximateIntegral(f, a, b, steps = 50):
    step_size = (b-a)/steps
    x = np.linspace(a, b, steps)
    y = f(x)
    return np.sum(y*step_size)

def plotFunction(x, y, ylim = 1):
    plt.plot(x,y, color='r')
    plt.xlim([x[0],x[-1]])
    plt.ylim([0, max(y) + ylim])
    plt.show()
    
def plotFunctions(title, functions, x = None, savefile=None, ylim=(0, 1.2)):
    if not x:
        x = np.linspace(-5, 5, 300)
    fig, ax = plt.subplots()
    
    legends = []
    colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
    color_index = 0
    for f in functions:
        legends.append("y=" + f[0])
        y = f[1](x)
        if(color_index == len(colors)):
            color_index = 0
        ax.plot(x,y,color=colors[color_index], alpha=0.5)
        color_index += 1
    ax.set(xlabel='x', ylabel='y', title=title)
    plt.xlim([x[0],x[-1]])
    plt.ylim([ylim[0], ylim[1]])
    ax.legend(legends)
    plt.grid(True)
    
    if savefile:
        plt.savefig(savefile + '.png')
    plt.show()
    
```

## Sigmoid and derivative
$$
\sigma(x) = \frac{1}{1+ e^{-z}} \\
\sigma'(x) = \sigma(x)(1 -\sigma(x))
$$

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def sigmoid_prime(x):
    sigmoid_x = sigmoid(x)
    return sigmoid_x * (1 - sigmoid(x))
```

```python
plotFunctions(r'Funciones sigmoides y derivadas', [(r'$\sigma(x)$', sigmoid), (r'$\sigma\'(x)$', sigmoid_prime)])
```

## Tanh and derivative
$$
tanh(x) = 2\sigma(2x) -1 = \frac{e^{x} - e^{-x}}{e^{x} + e^{-x}} \\
tanh'(x) = 1 - tanh(x)^2
$$

```python
def tanh(x):
    return 2*sigmoid(2*x) -1

def tanh_prime(x):
    return 1 - (tanh(x))**2

plotFunctions(r'Funciones Tanh y derivadas', [(r'$tanh$', tanh), (r'$tanh\'(x)$', tanh_prime)], ylim=(-1.2, 1.2))
```

## Hard Tanh

$$
hardtanh(x) = max(-1, min(1, x)) \\
hardtanh'(x) =
\begin{cases}
1 & -1 < x < 1 \\
0 & \text{otherwise}
\end{cases}
$$

```python
def hardtanh(x):
    return np.maximum(-1, np.minimum(1, x))

def hardtanh_prime(x):
    out = [1 if -1 < n and n < 1 else 0 for n in x]
    return np.array(out)
```

```python
plotFunctions(r'Hard Tanh y  derivative derivadas', [(r'$hard tanh$', hardtanh), (r'$hard tanh\'(x)$', hardtanh_prime)], ylim=(-1.2, 1.2))
```

```python
plotFunctions(r'derivada de tanh y derivada de sigmoide  ', [(r'$tanh\'(x)$', tanh_prime), (r'$\sigma\'(x)$', sigmoid_prime), (r'$hardtanh\'(x)$', hardtanh_prime)])
```

## ReLU and derivative
$$
ReLU(x) = max(0, x) \\
ReLU'(x) = 
\begin{cases}
0 & x \leq 0 \\
1 & 0 < x 
\end{cases}
$$

```python
def relu(x):
    return np.maximum(0, x)

def relu_prime(x):
    return x >= 0

plotFunctions(r'Relu y derivada', [(r'$ReLU(x)$', relu), (r'$ReLU\'(x)$', relu_prime)], ylim=(-0.2, 1.2))
```

## Leaky ReLU

$$
LeakyReLu(x) =\begin{cases}
x \hspace{0.3em} if \hspace{0.3em}x \geq 0 \\
\alpha x \hspace{0.3em} if \hspace{0.3em}x < 0 
\end{cases} \\
LeakyReLu'(x) =\begin{cases}
1 \hspace{0.3em} if \hspace{0.3em}x \geq 0 \\
\alpha \hspace{0.3em} if \hspace{0.3em}x < 0 
\end{cases}
$$

```python
def leaky_relu(x, alpha=0.05):
    return np.array([a if a >= 0 else alpha*a for a in x])

def leaky_relu_prime(x, alpha=0.05):
    return np.array([1 if a >= 0 else alpha for a in x])

plotFunctions(r'Leaky ReLU y derivada', [(r'$LReLU(x)$', leaky_relu), (r'$LReLU\'(x)$', leaky_relu_prime)], ylim=(-0.2, 1.2))
```

## ELU
Estas activaciones modifican la no linealidad para acercar las activaciones a 0, lo que podría acelerar la convergencia
$$
ELU(x) =\begin{cases}
                x \hspace{0.3em} if \hspace{0.3em}x > 0 \\
                \alpha (e^{x} - 1) \hspace{0.3em} if \hspace{0.3em}x \leq 0 
            \end{cases}
$$

```python
def elu(x, alpha=0.1):
    return np.array([a if a > 0 else alpha*(np.exp(a) -1) for a in x])

def elu_prime(x, alpha=0.1):
     return np.array([1 if a >= 0 else alpha*np.exp(a) for a in x])
    
plotFunctions(r'ELU and derivative functions', [(r'$ELU(x)$', elu), (r'$ELU\'(x)$', elu_prime)], ylim=(-0.2, 1.2))
```

## Softmax
$$
Softmax(x_i) = \frac{e^{x_{i}}}{\sum_{j} e^{x_{j}} }
$$

```python
def softmax(x):
    x_sum = np.sum(np.exp(x))
    return np.exp(x)/x_sum

print(softmax(np.array([4.0, 2.3, -1.7])))
```

## GELU
$$
GELU(x) = x P(X \leq x) = x \phi(x) = x . \frac{1}{2} \Big[1 + erf\Big(\frac{x}{\sqrt{2}}\Big)\Big]
$$
GELU can be approximated by:
$$
0.5x \Big[1 + tanh\Big( \sqrt{ \frac{2}{\pi} } . (x + 0.044715x^3) \Big) \Big]
$$
or
$$
x \sigma(1.702x)
$$

```python
def gelu(x):
    return 0.5*x * (1 + tanh( np.sqrt(2/np.pi) * (x + 0.044715 * np.power(x, 3))) )
```

```python
plotFunctions(r'Relu and GELU functions', [(r'$GELU(x)$', gelu), (r'$ReLU\'(x)$', relu)], ylim=(-0.2, 1.2))
```
